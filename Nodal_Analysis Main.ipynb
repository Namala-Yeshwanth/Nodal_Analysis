# Nodal Analysis of a Gas Well

This notebook performs **Nodal Analysis** for a gas well, generating Inflow Performance Relationship (IPR)
and Tubing Performance Relationship (TPR) curves for different tubing sizes. The analysis helps determine
the optimal operating point and assess how tubing diameter affects production.

## Contents
1. Data Preparation
2. IPR Curve Generation
3. TPR Curve Generation
4. Intersection Point Calculation (Operating Point)
5. Results and Conclusions

---


```python
   # import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib as inline   #inline makes sure that plot directly below the code cell
# if using .py script just use plt.show()

# plt.plot(x,y)
```

Now lets create IPR curve using Data flow rate vs Pwf

```python
# Lets get data for IPR (source: Petrowiki)
q_ipr = [0, 1999, 3094, 3902, 4512, 4963, 5275, 5458, 5519]
pwf_ipr = [4000, 3500, 3000, 2500, 2000, 1500, 1000, 500, 14.7]

ipr_df = pd.DataFrame({'Pwf':[0, 1999, 3094, 3902, 4512, 4963, 5275, 5458, 5519],
                  'Q': [4000, 3500, 3000, 2500, 2000, 1500, 1000, 500, 14.7]})

ipr = {
    'Pwf':pwf_ipr,
    'Q':q_ipr
}
ipr_df   #gives a table of q_ipr and pwf_ipr data
```

Now let's create IPR plot

```python
x_ipr=q_ipr; y_ipr=pwf_ipr

plt.figure(figsize=(8,4))
plt.xlabel('Qo (bbl/D)')
plt.ylabel('Pwf (psi)')
plt.legend('IPR Curve')
plt.plot(x_ipr, y_ipr)
```

```python
# Now Let's get TPR data
q_tpr=[1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000, 5500, 6000]

tpr_190 = [1334, 1400, 1487, 1592, 1712, 1843, 1984, 2132, 2287, 2446, 2609]
tpr_2375 = [1298, 1320, 1351, 1390, 1435, 1487, 1545, 1609, 1677, 1749, 1824]
tpr_2875 = [1286, 1294, 1305, 1319, 1336, 1356, 1378, 1403, 1431, 1461, 1493]

tpr = pd.DataFrame({'Q':q_tpr, 'Pwf(1.90" tbg)': tpr_190, 'Pwf(2.375" tbg)': tpr_2375, 'Pwf(2.875" tbg)': tpr_2875})
tpr
```

```python
# Let's plot TPR and IPR curves

plt.figure(figsize=(8,4))
plt.style.use('seaborn-v0_8-whitegrid')

plt.plot(ipr['Q'], ipr['Pwf'], label='IPR', linewidth=1.5)
plt.plot(tpr['Q'], tpr['Pwf(1.90" tbg)'], label='TPR for 1.90 inch Tubing', linewidth=1.5)
plt.plot(tpr['Q'], tpr['Pwf(2.375" tbg)'], label='TPR for 2.375 inch Tubing', linewidth=1.5)
plt.plot(tpr['Q'], tpr['Pwf(2.875" tbg)'], label='TPR for 2.875 inch Tubing', linewidth=1.5)

plt.xlabel('Qo (bbl/D)')
plt.ylabel('Pwf (psi)')
plt.title('Nodal Analysis for a Gas well')

plt.grid()
plt.legend(loc='best')

```

**Results:** The Nodal Analysis indicates that larger tubing diameters significantly improve production capacity by reducing flow resistance and associated pressure losses.
